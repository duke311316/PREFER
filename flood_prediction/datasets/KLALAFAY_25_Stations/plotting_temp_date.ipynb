```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Create a sample DataFrame with date, time, and temperature columns
data = {
    'Date': ['2023-08-01', '2023-08-02', '2023-08-03', '2023-08-04'],
    'Time': ['12:00:00', '14:00:00', '16:00:00', '18:00:00'],
    'Temperature': [25.5, 26.7, 28.3, 27.1]
}

df = pd.DataFrame(data)

# Combine 'Date' and 'Time' columns into a single 'Datetime' column
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
print(df)
# Plot the temperature against date and time
plt.figure(figsize=(10, 6))
plt.plot(df['Datetime'], df['Temperature'], marker='o')
plt.xlabel('Date and Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature vs. Date and Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()





```
